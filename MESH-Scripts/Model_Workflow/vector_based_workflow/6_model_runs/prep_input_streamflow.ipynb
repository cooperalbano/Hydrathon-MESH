```python
df1['DD'] = df1['DD'].astype(str)
df2['DD'] = df2['DD'].astype(str)
df1['YEAR'] = df1['YEAR'].astype(str)
df2['YEAR'] = df2['YEAR'].astype(str)
df1['YEAR-DAY'] = df1[['YEAR','DD']].apply('-'.join, axis = 1)
df2['YEAR-DAY'] = df2[['YEAR','DD']].apply('-'.join, axis = 1)
```
